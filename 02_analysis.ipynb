```python
import kagglehub
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
# read transformed data from data folder
data = pd.read_csv("data/cleaned_global_finance_data.csv")
```

```python
data.head()
```

```python
# barchart GDP growth
plt.figure(figsize=(12,6))
plt.bar(data['Country'], data['GDP_Growth_Rate_Percent'])
plt.xlabel('Country')
plt.xticks(rotation=45, ha='right')
plt.ylabel('GDP Growth (%)')
plt.title('GDP Growth Over Countries')
plt.show()
```

```python
# sort data by Market Cap
df_sorted = data.sort_values('Market_Cap_Trillion_USD', ascending=False)
# barchart market cap per country
plt.figure(figsize=(12,6))
plt.bar(df_sorted['Country'], df_sorted['Market_Cap_Trillion_USD'])
plt.xlabel('Country')
plt.xticks(rotation=45, ha='right')
plt.ylabel('Market Cap (Trillion USD)')
plt.title('Market Cap Over Countries')
plt.show()
```

```python
# sort data by Inflation Rate Percent
df_sorted = data.sort_values('Inflation_Rate_Percent', ascending=False)
# barchart inflation rate per country
plt.figure(figsize=(12,6))
plt.bar(df_sorted['Country'], df_sorted['Inflation_Rate_Percent'])
plt.xlabel('Country')
plt.xticks(rotation=45, ha='right')
plt.ylabel('Inflation Rate (%)')
plt.title('Inflation Rate Over Countries')
plt.show()
```
