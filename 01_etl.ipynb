# Extract Tranform and Load data

Dataset from **Kaggle**


```python
import kagglehub
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
# import dataset from kaggle
# Download latest version
path = kagglehub.dataset_download("imaadmahmood/global-finance-and-economic-indicators-dataset-2024")

print("Path to dataset files:", path)
```

```python
file_path = os.path.join(path, "Global finance data.csv")
file_path
```

```python
# read as csv with pandas
raw_data = pd.read_csv(file_path)
raw_data.head()
```

```python
raw_data.tail()
```

```python
# check for NaN values
raw_data.isna().sum()
```

```python
# basic plot if GDP growth over years
plt.figure(figsize=(12,6))
plt.bar(raw_data['Country'], raw_data['GDP_Growth_Rate_Percent'])
plt.xlabel('Year')
plt.ylabel('GDP Growth (%)')
plt.title('GDP Growth Over Years')
plt.show()
```
