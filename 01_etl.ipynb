```python
import kagglehub
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
# import dataset from kaggle
# Download latest version
path = kagglehub.dataset_download("imaadmahmood/global-finance-and-economic-indicators-dataset-2024")

print("Path to dataset files:", path)
```

```python
file_path = os.path.join(path, "Global finance data.csv")
file_path
```

```python
# read as csv with pandas
raw_data = pd.read_csv(file_path)
raw_data.head()
```

```python
raw_data.tail()
```

```python
# Checks types and non-null counts
raw_data.info()
```

```python
# Checks distribution of numerical data
raw_data.describe()     
```

```python
# Counts duplicate rows
raw_data.duplicated().sum() 
```

```python
# check for NaN values
raw_data.isna().sum()
```

## Export as CSV File

save to /data folder

```python
# check for data folder, create one if it doesn't exist
if not os.path.exists("data"):
    os.makedirs("data")
```

```python
# export as csv
raw_data.to_csv("data/cleaned_global_finance_data.csv", index=False)
```
